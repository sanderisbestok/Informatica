```python
# Generates a random int
def randint(a, c, m):
    global seed   
    seed = np.mod((a * seed + c), m)
    
    return seed

# Generates a random number between 0,1
def rand():
    number = randint(1664525, 1013904223, (2**(32)))
    number_float = float(number / (2**(32)))
    
    return number_float

# Inverse CDF of lognormal
def icdf(mu, sigma, p):
    return np.exp(sp.erfinv(2*p-1)*sigma*np.sqrt(2)+mu)

# Gives a random number with the log-normal distribution
def randlognorm(mu, sigma):
    return icdf(mu, sigma, rand())
```

### Question 2 & Question 3

```python
def percentile(values, alpha):
    values = np.sort(values)
    k = alpha * (len(values) - 1)
    kceil = math.ceil(k)
    kfloor = math.floor(k)
    
    return ((values[kfloor] * (kceil - k)) + (values[kceil] * (k - kfloor))) 

print("0,95-Percentile of {1,4,5}: %.1f" % percentile([1,4,5], 0.95))
```

```python
mu = 0
sigma = 0.3

seed = 0
n = 1000

x = [randlognorm(mu, sigma) for x in range(0, n)]
```

```python
def lognormalpdf(mu, sigma, x):
    return (1/x)*(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-((np.log(x)-mu)**2)/(2*sigma**2))
```

```python
plt.figure(figsize=(10,8))
_, bins, _ = plt.hist(x, 100, normed=True)
plt.plot(bins, lognormalpdf(mu, sigma, bins), label='Lognormal PDF')
plt.axvline(x = percentile(x, 0.25), label='0,25-Percentile', color='r')
plt.axvline(x = percentile(x, 0.50), label='0,50-Percentile', color='g')
plt.axvline(x = percentile(x, 0.75), label='0,75-Percentile', color='b')
plt.legend()
plt.show()
```

### Question 3

```python
n = 100
N = 100000

x = [randlognorm(mu, sigma) for x in range(0, n)]

bootstrap = [calc_mean(np.random.choice(x, n, replace=True)) for i in range(N)]
theoretical = [calc_mean([randlognorm(mu, sigma) for x in range(n)]) for i in range(N)]
    
```

```python
def pdf(mu, sigma, x):
    return (1 / (sigma * ((2 * math.pi) ** 0.5))) * (math.e ** ((-0.5) * (((x - mu) / sigma) ** 2)))
```

```python
mean_b = calc_mean(bootstrap)
stdev_b = standard_deviation(bootstrap)

x = np.linspace(0.9, 1.2, 100)
y = pdf(mean_b, stdev_b, x)
    

plt.axvline(x = mean_b - 2 * stdev_b, label='Lower', color='r')
plt.axvline(x = mean_b + 2 * stdev_b, label='Upper', color='g')
plt.axvline(x = mean_b, label='Mean', color='b')
plt.plot(x, y, label="Fitted")
plt.legend()

plt.hist(bootstrap, 100, normed=True)
plt.show()
```

```python
mean_t = calc_mean(theoretical)
stdev_t = standard_deviation(theoretical)

plt.axvline(x = mean_t - 2 * stdev_t, label='Lower', color='r')
plt.axvline(x = mean_t + 2 * stdev_t, label='Upper', color='g')
plt.axvline(x = mean_t, label='Mean', color='b')
plt.legend()

plt.hist(theoretical, 100, normed=True)
plt.show()
```
