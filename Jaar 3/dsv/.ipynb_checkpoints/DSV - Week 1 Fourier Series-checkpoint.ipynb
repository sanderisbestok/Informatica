```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

First we will plot and configure the original function

```python
def y(t):
    return np.sin(np.pi*t) + np.sin(2*np.pi*t) + np.sin(5 * np.pi * t)

t = np.linspace(0, 2, 200)
original = y(t)
```

```python
plt.figure(figsize=(10,10))
plt.title("Original function")
plt.xlabel("t")
plt.ylabel("y")
plt.plot(t, original)
plt.show()
```

Now we can write the functions for the fourier series.

```python
# To make integrate easier in python we write y(t) in a_n
def ya(t, n, P):
    return (np.sin(np.pi*t) + np.sin(2*np.pi*t) + np.sin(5 * np.pi * t)) * np.cos((2*np.pi*n*t)/P)
    
def yb(t, n, P):
    return (np.sin(np.pi*t) + np.sin(2*np.pi*t) + np.sin(5 * np.pi * t)) * np.sin((2*np.pi*n*t)/P)
    

def S(t, N, P):
    term1 = integrate.quad(ya, 0, P, args=(0, P))[0]
    term2 = 0
    term3 = 0

    for n in range(1, N):
        term2 += 2/P * integrate.quad(ya, 0, P, args=(n, P))[0] * np.cos((2*np.pi*n*t)/P)
        term3 += 2/P * integrate.quad(yb, 0, P, args=(n, P))[0] * np.sin((2*np.pi*n*t)/P)
        
    return term1 + term2 + term3
    
approx3 = S(t, 3, 2)
approx5 = S(t, 5, 2)
approx8 = S(t, 8, 2)
```

```python
plt.figure(figsize=(7,7))
plt.title("Original function and fourier n=3")
plt.xlabel("t")
plt.ylabel("y")
plt.plot(t, original, label="Original")
plt.plot(t, approx3, label="Fourier with n=3")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.title("Original function and fourier n=5")
plt.xlabel("t")
plt.ylabel("y")
plt.plot(t, original, label="Original")
plt.plot(t, approx5, label="Fourier with n=5")
plt.legend()
plt.show()
```
