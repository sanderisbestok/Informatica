```python
t = np.linspace(-0.02, 0.05, 1000)
plt.subplot(2,1,1); plt.plot(t, np.exp(2j*np.pi*25*t).real );
plt.xlabel('t');
plt.ylabel('Re x(t)');
plt.title(r'Real part of  \$x(t)=e^{j 25 \pi t}\$');
plt.xlim([-0.02, 0.05]);
plt.subplot(2,1,2); plt.plot(t, np.exp(2j*np.pi*25*t).imag);
plt.xlabel('t');
plt.ylabel('Im x(t)');
plt.title(r'Imaginary part of  \$x(t)=e^{j 25\pi t}\$');
plt.xlim([-0.02, 0.05]);
#@savefig csineplot.png
plt.show()
```

## Assignment 4

The formula for converting cartesian to polar goes as follows:
$$z=r(\cos(\phi)+j\sin(\phi))$$
$$r = \sqrt(a^2+b^2)$$
$$\phi = \tan^-(\frac b a)$$

```python
z1 = 4+3j
z2 = -3+5j
z3 = 6-2j
z4 = -2+4j



def polar(z):
    r = np.sqrt(z.real**2 + z.imag**2)
    phi = math.atan2(z.imag,z.real)
    return r, phi

print("r = %f, phi = %f" % polar(z1))
print("r = %f, phi = %f" % polar(z2))
print("r = %f, phi = %f" % polar(z3))
print("r = %f, phi = %f" % polar(z4))
```

```python
x = np.linspace(0, 0.5, 40)
plt.plot(x, x, label="$x$")
plt.plot(x, np.sin(x), label="$sin(x)$")
plt.plot(x, np.cos(x), label="$cos(x)$")
plt.plot(x, 1-((x**2)/2), label="$1-(x^2/2)$")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 40)
plt.plot(x, x, label="$x$")
plt.plot(x, np.sin(x), label="$sin(x)$")
plt.plot(x, np.cos(x), label="$cos(x)$")
plt.plot(x, 1-((x**2)/2), label="$1-(x^2/2)$")
plt.legend()
plt.show()
```

The bigger values we use the more they deviate.

```python
x = np.linspace(0, 0.5, 40)
plt.plot(x, np.exp(x), label="$e^x$")
plt.plot(x, (1+x+(x**2/2)), label="$1+x+(x^2/2$")
plt.legend()
plt.show()
```
