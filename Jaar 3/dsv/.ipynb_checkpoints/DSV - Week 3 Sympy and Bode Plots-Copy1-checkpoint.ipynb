```python
from sympy.plotting import plot
from sympy.abc import symbols
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
sp.init_printing(use_unicode=True)
```

## Assignment 1

```python
from sympy import *

x = symbols('x')
print("Integral of sin(x**2) =")
integrate(sin(x**2), (x, -sp.oo, sp.oo))
```

## Assignment 2

```python
k = symbols('k')

print("Result")
print(fourier_transform(1, x, k))
print("Expected (because of the area)")
print("1")
print("Heaviside")
print(fourier_transform(Heaviside(x), x, k))
print("Expected (because of the area)")
print("1/2")
```

## Assignment 3

```python
plot(sp.Heaviside(x), (x, -np.pi, np.pi))
plot(fourier_series(sp.Heaviside(x), (x, -np.pi, np.pi)).truncate(100), (x, -np.pi, np.pi))
plot(0.5*x, (x, -np.pi, np.pi))
plot(fourier_series(0.5*x, (x, -np.pi, np.pi)).truncate(100), (x, -np.pi, np.pi))
plot(DiracDelta(x), (x, -np.pi, np.pi))
plot(fourier_series(DiracDelta(x), (x, -np.pi, np.pi)).truncate(100), (x, -np.pi, np.pi))
```

## Assignment 4

```python
w = symbols('w')
H = (1/(1+I*(w/1000)))

plot(abs(H), (w, 1, 5000), xscale='log')
plot(arg(H), (w, 1, 5000), xscale='log')
```

The slope of the filter is -20 db/decade

### Assignment 5

```python
H = lambda n: 1/(1+((I*w)/(I*1000))**2*n)

plot(H(0), (w, 1, 5000), xscale='log')
plot(H(1), (w, 1, 5000), xscale='log')
plot(H(2), (w, 1, 5000), xscale='log')
plot(H(4), (w, 1, 5000), xscale='log')
```

The slope is -20n db/decade
