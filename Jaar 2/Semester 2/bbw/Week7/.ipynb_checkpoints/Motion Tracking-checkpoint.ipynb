First we implemented the motion tracking with the help of a function of skimage which will match a template with the image and return the normalized cross corelation.

```python
image = reader.get_data(40)
template = image[275:325, 75:120]

template = skimage.color.rgb2gray(template)
image = skimage.color.rgb2gray(image)

plt.figure(figsize=(10, 2))
plt.subplot(1, 3, 1)
plt.title("Frame")
plt.imshow(image, cmap=plt.cm.gray)
plt.subplot(1, 3, 3)
plt.title("Template")
plt.imshow(template, cmap=plt.cm.gray)

result = match_template(image, template)
plt.subplot(1, 3, 2)
plt.title("nnc")
plt.imshow(result)
plt.show()
```

We can get the locations by applying an argmax on the resulting image of every frame.

```python
locations = []

for i in range(40,350):
    im = reader.get_data(i)
    image = skimage.color.rgb2gray(im)
    
    result = match_template(image, template)
    locations.append(argmax(result))
```

Then we can make a scatter plot with the locations. 

Here the last frame shown with this scatter plot.

```python
index = np.unravel_index(locations, result.shape)

plt.imshow(reader.get_data(350))
plt.scatter(index[1] + 20 , index[0] + 20)
plt.show()
```
