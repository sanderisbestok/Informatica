# Camera Calibration

In this excercise we need to calibrate a camera given a certain set of points. Therefor we need to calculate the camera matrix.

## Theory
We have to find the relation between the 3D and the projected 2D point. This can be calculated with the matrix $P$.

The 3D point $\widetilde{X}$ and the 2D point $\widetilde{x}$ have the following relation:

$$\widetilde{x}~P\widetilde{X}$$

Of course $\widetilde{x} will be homogeneous. We can calculate the cross product where $p_i$ are the row vectors of $P$:

$$
\widetilde{x} \times P \widetilde{X} =
 \begin{pmatrix}
   x\widetilde{X}^T\widetilde{p_3}  - \widetilde{X}^T\widetilde{p_1} \\
   y\widetilde{X}^T\widetilde{p_3} - \widetilde{X}^T\widetilde{p_2} \\
   x\widetilde{X}^T \widetilde{p_2} - y \widetilde{X}^T\widetilde{p_1}
 \end{pmatrix}
 = \begin{pmatrix}
     0\\
     0\\
     0
    \end{pmatrix}
$$

Because the third element in the vector is just a combination of the first two we can rewrite it to a simpeler form:

$$\begin{pmatrix}
    \widetilde{X}^T & 0^T & -x\widetilde{X}^T\\
    0^T & \widetilde{X}^T & -y\widetilde{X}^T
    \end{pmatrix}
   p = \begin{pmatrix}
     0\\
     0\\
     0
    \end{pmatrix}
$$

This is just for on point, it can be stacked with 2n equations on top of each other which will give us more points. We will denote this as

$$Ap = 0$$

But this method will only work if there are no measurement errors at all. Because we measure the imagegs there will be some noise. So we will have to find a vector that minimize $||Ap||$

This can be done with the so called SVD trick we used before. Afterwards the vector $p$ can be reshaped in camera matrix P.

Then we needed a function to calibrate the camera given the corresponding point sets we just loaded in. The return should be matrix $P$. $P$ will project points $(X,Y,Z)$ onto the 2D points $(x,y)$

This has been done using the method described in the theory. The algorithm itself is quite straight forward;

```python
def calibrate_cam(xy, xyz):
    """
    Use the SVD trick to generate a camera matrix P 
        - xy is a list of 2D coordinates
        - xyz is a list of 3D coordinates
    """
    #We use a base of zeros to be able to vstack onto that
    A = zeros(12)
    for idx, (x, y, z) in enumerate(xyz):
        
        #For every 3D point, create 2 rows in matrix A which will be used in our SVD trick
        x2, y2 = xy[idx]
        r1 = array([x, y, z, 1,  0, 0, 0, 0, -x2 * x, -x2 * y, -x2 * z, -x2])
        r2 = array([0, 0, 0, 0, x, y, z, 1, -y2 * x, -y2 * y, -y2 * z, -y2])
        A = vstack((A, r1))
        A = vstack((A, r2))
    
    #Call SVD with matrix A without the row of zeros
    _, _, V = linalg.svd(A[1:])
    p = V[-1]
    return p.reshape(3, 4)
```

## Testing

The next step is to test our matrix P. The simplest way to do this is to check if the 3D points we have correspondent with the right 2D points after we applied the matrix.

We will calculate a so called reporjection error. Which is the average euclidean distance error of the 2D points.

```python
#Calculate the camera matrix P and convert the 3D coordinates to Homogeneous coordinates by adding a 1 to
# every row
P  = calibrate_cam(xy, XYZ)
homogeneous = 18 * [[1]]
coordinates = append(XYZ, homogeneous, axis=1)
dist = 0

# Loop through all the 3D coordinates and calculate the coordinates of the projection of this point
for idx, coord in enumerate(coordinates):
    result = P @ coord
    point = result / result[2]
    plot(point[0], point[1], 'd', color ='cyan')
    #Calculate euclidean distance
    dist += numpy.linalg.norm(point[:2]-xy[idx])

#Calculate reprojection error and print image
#Reprojection error = Mean of all euclidean distances
dist = dist / len(xy) 
print('The reprojection error is: '  + str(dist))
axis('off')
title('Projection of the 3D points from XYZ')
imshow(im)
show()
```

Another test we can do is making 3D unit cubes and place them at a given location in 3D space with the help of $P$.

```python
import itertools


def drawCube(P, X, Y, Z):
    """
    Plot a cube with the 3D point as one of the bottom corners
    """
    corners = array([[X, Y, Z, 1] , [X, Y+1, Z, 1] , [X, Y+1, Z+1, 1], [X, Y, Z + 1, 1], [X + 1, Y, Z + 1, 1], 
                    [X + 1, Y, Z, 1] , [X + 1 , Y + 1, Z + 1, 1], [X + 1, Y + 1, Z, 1]])
    
    # Check for all possible combinations between the cornerpoints of our cube and only connect those
    # where the euclidean distance is 1.0
    for p1, p2 in itertools.product(corners, corners):
        if norm(p1 - p2) == 1.0:
            coord1 = P @ p1
            coord1 = coord1 / coord1[2]
            coord2 = P @ p2
            coord2 = coord2 / coord2[2]
            plot([coord1[0], coord2[0]], [coord1[1], coord2[1]], color='cyan')


axis('off')
imshow(im)
    
drawCube(P, -5, 0, 0)
drawCube(P, 0, 0, 0)
drawCube(P, 0, -5, 5)
drawCube(P, 0, -2, 3)
drawCube(P, -7, 0, 5)
show()
```
